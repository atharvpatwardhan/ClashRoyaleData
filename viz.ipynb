```python
arenas = battle_data["arena.name"].unique().tolist()

arenas.sort()
# Create a selector widget for arenas
arena_selector = pn.widgets.Select(name="Arena", options=arenas)
```

```python
# Function to filter and summarize data by arena
def get_card_usage(arena):
    filtered_battle_data = battle_data[battle_data["arena.name"] == arena]

    # Summarize card counts for winners and losers
    winner_card_counts = {
        "Common": filtered_battle_data["winner.common.count"].sum(),
        "Rare": filtered_battle_data["winner.rare.count"].sum(),
        "Epic": filtered_battle_data["winner.epic.count"].sum(),
        "Legendary": filtered_battle_data["winner.legendary.count"].sum(),
    }

    loser_card_counts = {
        "Common": filtered_battle_data["loser.common.count"].sum(),
        "Rare": filtered_battle_data["loser.rare.count"].sum(),
        "Epic": filtered_battle_data["loser.epic.count"].sum(),
        "Legendary": filtered_battle_data["loser.legendary.count"].sum(),
    }

    # Prepare data for plotting
    data = {
        "Card Type": list(winner_card_counts.keys()) * 2,
        "Count": list(winner_card_counts.values()) + list(loser_card_counts.values()),
        "Category": ["Winner"] * 4 + ["Loser"] * 4,
    }

    return pd.DataFrame(data)
```

```python
def get_average_elixir(arena):
    print(arena)
    filtered_battle_data = battle_data[battle_data["arena.name"] == arena]
    data = {
        "winner_e_mean_mean" : filtered_battle_data["winner.elixir.average"].mean(),
        "winner_e_mean_modes" : filtered_battle_data["winner.elixir.average"].mode(),
        "winner_e_mean_median" : filtered_battle_data["winner.elixir.average"].median(),
        "loser_e_mean_mean" : filtered_battle_data["loser.elixir.average"].mean(), 
        "loser_e_mean_modes" : filtered_battle_data["loser.elixir.average"].mode(), 
        "loser_e_mean_median" : filtered_battle_data["loser.elixir.average"].median(),
        "l" : (filtered_battle_data["loser.elixir.average"]<filtered_battle_data["winner.elixir.average"]).sum(),
        "w" : (filtered_battle_data["loser.elixir.average"]>filtered_battle_data["winner.elixir.average"]).sum()
    }
    
    return pd.DataFrame(data)
    
```

```python
def spells(arena):
    print(arena)
    filtered_battle_data = battle_data[battle_data["arena.name"] == arena]
    print("total battles ",len(filtered_battle_data))
    print("mean spell count",filtered_battle_data["winner.spell.count"].mean())
    count_w = (filtered_battle_data['winner.spell.count']>0).sum()
    print("ratio of winners using spells", count_w/len(filtered_battle_data))
    count_wl = ((filtered_battle_data['winner.spell.count']>0) & (filtered_battle_data['loser.spell.count']==0)).sum()
    print("used only by winner",count_wl)
    count_lw = ((filtered_battle_data['winner.spell.count']==0) & (filtered_battle_data['loser.spell.count']>0)).sum()
    print("used only by loser",count_lw)
    count_l = (filtered_battle_data['loser.spell.count']>0).sum()
    print("ratio of losers using spells", count_l/len(filtered_battle_data))
```

```python
@pn.depends(arena_selector)
def update_plot(arena):
    plot_battle_data = get_card_usage(arena)
    bars = hv.Bars(plot_battle_data, ["Card Type", "Category"], "Count").opts(
        opts.Bars(
            width=700,
            height=400,
            title=f"Card Usage by Category in {arena}",
            xlabel="Card Type",
            ylabel="Count",
            legend_position="top_right",
            tools=["hover"],
            color=hv.dim("Category").categorize({"Winner": "blue", "Loser": "red"}),
            show_legend=True,
        )
    )
    return bars
```

```python
@pn.depends(arena_selector)
def get_top_cards_by_arena(arena):
    top_cards = battle_data_grand_totals[battle_data_grand_totals["arena.name"] == arena]
    card_usage = top_cards.drop(columns=['arena.name'])
    total_usage = card_usage.sum().sort_values(ascending=False)
    top_10_cards = total_usage.head(10)
    
    bar_plot = hv.Bars(top_10_cards)
    
    bar_plot.opts(
    title= f'Top 10 Most Used Cards in Arena {arena}',
    xlabel="Card",
    ylabel="Usage",
    xrotation=45,
    width=800,
    height=400,
    color="blue",
)

    return bar_plot
```

```python
dashboard = pn.Column(
    pn.pane.Markdown("Clash Royale Analytics Dashboard"),
    arena_selector,
    update_plot,
    get_top_cards_by_arena
)

# Serve the dashboard
dashboard.show()
```
