```python
def arena_placer(row):
    arena_name = None
    
    if(row['average.startingTrophies'] < 300):
        arena_name = '1: Goblin Stadium'
    elif ((row['average.startingTrophies'] >= 300)&(row['average.startingTrophies'] < 600)):
        arena_name = '2: Bone Pit' 
    elif ((row['average.startingTrophies'] >= 600)&(row['average.startingTrophies'] < 1000)):
        arena_name = '3: Barbarian Bowl'
    elif ((row['average.startingTrophies'] >= 1000)&(row['average.startingTrophies'] < 1300)):
        arena_name = '4: Spell Valley'
    elif ((row['average.startingTrophies'] >= 1300)&(row['average.startingTrophies'] < 1600)):
        arena_name = '5: Builder\'s Workshop'
    elif ((row['average.startingTrophies'] >= 1600)&(row['average.startingTrophies'] < 2000)):
        arena_name = '6: P.E.K.K.A\'s Playhouse'
    elif ((row['average.startingTrophies'] >= 2000)&(row['average.startingTrophies'] < 2300)):
        arena_name = '7: Royal Arena'
    elif ((row['average.startingTrophies'] >= 2300)&(row['average.startingTrophies'] < 2600)):
        arena_name = '8: Frozen Peak'  
    elif ((row['average.startingTrophies'] >= 2600)&(row['average.startingTrophies'] < 3000)):
        arena_name = '9: Jungle Arena'
    elif ((row['average.startingTrophies'] >= 3000)&(row['average.startingTrophies'] < 3400)):
        arena_name = '10: Hog Mountain'
    elif ((row['average.startingTrophies'] >= 3400)&(row['average.startingTrophies'] < 3800)):
        arena_name = '11: Electro Valley'  
    elif ((row['average.startingTrophies'] >= 3800)&(row['average.startingTrophies'] < 4200)):
        arena_name = '12: Spooky Town'
    elif ((row['average.startingTrophies'] >= 4200)&(row['average.startingTrophies'] < 4600)):
        arena_name = '13: Rascal\'s Hideout'
    elif ((row['average.startingTrophies'] >= 4600)&(row['average.startingTrophies'] < 5000)):
        arena_name = '14: Serenity Peak'
    elif ((row['average.startingTrophies'] >= 5000)&(row['average.startingTrophies'] < 5500)):
        arena_name = '15: Miner\'s Mine'
    elif ((row['average.startingTrophies'] >= 5500)&(row['average.startingTrophies'] < 6000)):
        arena_name = '16: Executioner\'s Kitchen'
    elif ((row['average.startingTrophies'] >= 6000)&(row['average.startingTrophies'] < 6500)):
        arena_name = '17: Royal Crypt'
    elif ((row['average.startingTrophies'] >= 6500)&(row['average.startingTrophies'] < 7000)):
        arena_name = '18: Silent Sanctuary'
    elif ((row['average.startingTrophies'] >= 7000)&(row['average.startingTrophies'] < 7500)):
        arena_name = '19: Dragon Spa'
    elif ((row['average.startingTrophies'] >= 7500)&(row['average.startingTrophies'] < 8000)):
        arena_name = '20: Boot Camp'
    elif ((row['average.startingTrophies'] >= 8000)&(row['average.startingTrophies'] < 8500)):
        arena_name = '21: Clash Fest'    
    elif ((row['average.startingTrophies'] >= 8500)&(row['average.startingTrophies'] < 9000)):
        arena_name = '22: PANCAKES!'    
    elif(row['average.startingTrophies'] >= 9000):
        arena_name = '23: Legendary Arena' 
        
    return arena_name

df['average.startingTrophies'] = df['average.startingTrophies'].fillna(0)
# df['arena.name'] = None
df['arena.name'] = df.apply(arena_placer, axis=1)
```

```python
arenas = df["arena.name"].unique().tolist()

arenas.sort()
# Create a selector widget for arenas
arena_selector = pn.widgets.Select(name="Arena", options=arenas)
```

```python
# Function to filter and summarize data by arena
def get_card_usage(arena):
    filtered_df = df[df["arena.name"] == arena]

    # Summarize card counts for winners and losers
    winner_card_counts = {
        "Common": filtered_df["winner.common.count"].sum(),
        "Rare": filtered_df["winner.rare.count"].sum(),
        "Epic": filtered_df["winner.epic.count"].sum(),
        "Legendary": filtered_df["winner.legendary.count"].sum(),
    }

    loser_card_counts = {
        "Common": filtered_df["loser.common.count"].sum(),
        "Rare": filtered_df["loser.rare.count"].sum(),
        "Epic": filtered_df["loser.epic.count"].sum(),
        "Legendary": filtered_df["loser.legendary.count"].sum(),
    }

    # Prepare data for plotting
    data = {
        "Card Type": list(winner_card_counts.keys()) * 2,
        "Count": list(winner_card_counts.values()) + list(loser_card_counts.values()),
        "Category": ["Winner"] * 4 + ["Loser"] * 4,
    }

    return pd.DataFrame(data)
```

```python
@pn.depends(arena_selector)
def update_plot(arena):
    plot_df = get_card_usage(arena)
    bars = hv.Bars(plot_df, ["Card Type", "Category"], "Count").opts(
        opts.Bars(
            width=700,
            height=400,
            title=f"Card Usage by Category in {arena}",
            xlabel="Card Type",
            ylabel="Count",
            legend_position="top_right",
            tools=["hover"],
            color=hv.dim("Category").categorize({"Winner": "blue", "Loser": "red"}),
            show_legend=True,
        )
    )
    return bars
```

```python
df_winner_combined = df_winner_card1 + df_winner_card2 + df_winner_card3 + df_winner_card4 + df_winner_card5 + df_winner_card6 + df_winner_card7 + df_winner_card8
df_loser_combined = df_loser_card1 + df_loser_card2 + df_loser_card3 + df_loser_card4 + df_loser_card5 + df_loser_card6 + df_loser_card7 + df_loser_card8
df_grand_totals = df_winner_combined + df_loser_combined
```

```python
df_cardmasterlist = pd.read_csv("D:\\Data-Viz-Project\\CardMasterListSeason18_12082020.csv")
df_cardmasterlist.reset_index(inplace=True)
cardmasterlist_dict = dict(zip(df_cardmasterlist["team.card1.id"],df_cardmasterlist["team.card1.name"]))
```

```python
df_grand_totals = df_grand_totals.rename(columns=cardmasterlist_dict)
```

```python
df_grand_totals.head()
```

```python
df_grand_totals.columns = df_grand_totals.columns.str.strip()
```

```python
@pn.depends(arena_selector)
def get_top_cards_by_arena(arena):
    top_cards = df_grand_totals[df_grand_totals["arena.name"] == arena]
    card_usage = top_cards.drop(columns=['arena.name'])
    total_usage = card_usage.sum().sort_values(ascending=False)
    top_10_cards = total_usage.head(10)
    
    bar_plot = hv.Bars(top_10_cards)
    
    bar_plot.opts(
    title= f'Top 10 Most Used Cards in Arena {arena}',
    xlabel="Card",
    ylabel="Usage",
    xrotation=45,
    width=800,
    height=400,
    color="blue",
)

    return bar_plot
```

```python
dashboard = pn.Column(
    pn.pane.Markdown("Clash Royale Analytics Dashboard"),
    arena_selector,
    update_plot,
    get_top_cards_by_arena
)

# Serve the dashboard
dashboard.show()
```
